# Simulation of Modified Kicked Rotor Hamiltonian System

The classic Hamiltonian system, perhaps the archetype of all Hamiltonian systems, is the harmonic oscillator. When pulsed periodically and mapped onto the Poincaré plane, such a simple dynamical system creates magnificent tapestries.

Reference: David D. Nolte, [Galileo Unbound](https://galileo-unbound.blog/2018/10/27/how-to-weave-a-tapestry-from-hamiltonian-chaos/) 

```python
import numpy as np
from scipy import integrate
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = 12,12
np.random.seed(2)
```

```python

def tapestry(q, K=None, n=5):
    """
    Periodically- Kicked Hamiltonian mapped onto Poincaré plane.
    
    Parameters
    ----------
    q: periodicity
        If q is a small even number, tapestries of regular symmetric are generated.
        If it is odd, the tapestries turn into quasi-crystals.
    
    K: the coupling constant
        For small K, the tapestries are mostly regular. 
        But as the coupling K increases, stochastic layers emerge.
    """
    
    phi = (1+np.sqrt(n))/2
    if K is None: K = 1 - phi 
    alpha = 2*np.pi/q
    #plt.figure();
    plt.figure(facecolor = 'mistyrose')
    for eloop in range(0,1000):
        xlast = 50*np.random.random()
        ylast = 50*np.random.random()
    
        xnew = np.zeros(shape=(300,))
        ynew = np.zeros(shape=(300,))
        for loop in range(0,300):
            xnew[loop] = (xlast + K*np.sin(ylast))*np.cos(alpha) + ylast*np.sin(alpha)
            ynew[loop] = -(xlast + K*np.sin(ylast))*np.sin(alpha) + ylast*np.cos(alpha)
            xlast = xnew[loop]
            ylast = ynew[loop]

        plt.plot(np.real(xnew),np.real(ynew),'o',ms=1)
        plt.axis('off');
```

```python
tapestry(4)
```

```python
tapestry(7, .656)
```

```python
tapestry(5, 0.052)
```
