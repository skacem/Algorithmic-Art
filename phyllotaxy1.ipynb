```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
# Set up Notebook
%matplotlib inline

# Disable jedi autocompleter
%config Completer.use_jedi = False

# set notebook environment
%precision %.3f
pd.options.display.float_format = '{:,.3f}'.format
plt.rcParams["figure.figsize"] = 10,10
pd.set_option('display.width', 100)
plt.rcParams.update({'font.size': 14})
```

```python
t = np.linspace(start=0, stop=2*np.pi, num=50)
x = np.sin(t)
y = np.cos(t)
```

```python
df = pd.DataFrame(np.array([t, x, y]).T, columns=['t', 'x', 'y'])
```

```python
plt.scatter(df.x, df.y, s=120);
plt.xlabel('X')
plt.ylabel('Y')
```

Plants arrange their leaves in spirals. So let's plot a spiral. A simple way to arrange our points from the above figure into a spiral is to multiply x and y by a factor which increases for each point. we could use t as that factor:

```python
plt.scatter(df.x*t, df.y*t, s=120);
plt.xlabel('X')
plt.ylabel('Y')
```

Not so good, right? 
So let's do something more fancy. We can use the harmonical property of golden angle:
$$ angle = \pi\times(3 - \sqrt{5})$$

The golden angle is inspired by the [Golden Ratio](https://en.wikipedia.org/wiki/Golden_ratio). Both the Golden Ratio and the Golden Angle appear in unexpected places in nature. Apart of flower petals and plant leaves, you'll find them in seed heads, pine cones, sunflower seeds, shells, spiral galaxies, hurricanes, etc.

```python
angle = np.pi * (3 - np.sqrt(5))
t = np.arange(500) * angle
x = np.sin(t)
y = np.cos(t)
plt.scatter(x*t, y*t, s=80);
```

Let's make it looks more like a plant

```python
plt.scatter(x*t, y*t, s=1600, c='forestgreen', alpha=.6, edgecolors='darkgreen')
plt.axis('off');
```

et voila!

We can make it more spacy, just by changing the marker and the background color:

```python
fig = plt.scatter(x*t, y*t, s=999, marker='^', c='gold', alpha=.4, linewidth=3)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
fig.axes.set_facecolor('darkmagenta')
```

what if we change the angle:


```python
angle = 11 * np.pi / 180
t = np.arange(-1200, 1199) * angle
x = np.sin(t)
y = np.cos(t)
fig = plt.scatter(x*t, y*t, s=129, alpha=.3, c='yellow')
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
fig.axes.set_facecolor('lightcoral');
```
