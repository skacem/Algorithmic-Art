# Multi-Pendulum Homograph
A harmonograph is a type of drawing machine that creates harmonic geometric illustrations. There are a few kinds of harmonographs, we'll be implementing a multi-pendulum homograph.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.rcParams["figure.figsize"] = 10,8
# Disable jedi autocompleter
%config Completer.use_jedi = False
```

```python
def harmonograph(pendulum, frequency, sigma = 0.002):
    """
    Multi-pendulum random Harmonograph simulator.

    :praram pendulum: Number of pendulum to simulate
    :type pendulum: int
    :param frequency: The frequency of the pendulum
    :type frequency: int
    :param sigma: The variance or spread from the given frequency. 
    :type sigma: float
    """


    sigma = 0.002
    # step size
    step = 0.01
    # of steps
    steps = 40000

    # time axis
    t = np.arange(steps)*step      
    # decay vector
    d = 1 - np.arange(steps)/steps 


    ax = [random.uniform(0, 1) for i in range(pendulum)]
    ay = [random.uniform(0, 1) for i in range(pendulum)]
    px = [random.uniform(0, 2*np.pi) for i in range(pendulum)]
    py = [random.uniform(0, 2*np.pi) for i in range(pendulum)]

    fx = [random.randint(1, frequency) + random.gauss(0, sigma) for i in range(pendulum)]
    fy = [random.randint(1, frequency) + random.gauss(0, sigma) for i in range(pendulum)]

    # Polar to Cartesian:
    x = y = 0
    for i in range(pendulum):
        x += d * (ax[i] * np.sin(t * fx[i] + px[i]))
        y += d * (ay[i] * np.sin(t * fy[i] + py[i]))

    plt.figure(facecolor = 'floralwhite')
    plt.plot(x, y, 'midnightblue', linewidth=1, alpha=.8)
    plt.axis('off')
    plt.subplots_adjust(left=0.0, right=1.0, top=1.0, bottom=0.0);
```

```python
harmonograph(5, 3)
```

```python
harmonograph(4, 4, 0.9)
```

```python
harmonograph(2, 2, 0.5)
```
